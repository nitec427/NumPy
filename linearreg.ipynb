```python
A_M = np.log(2) / 2
B_M = np.log(2250) - A_M * 1971
Mooreslaw = lambda year: np.exp(B_M) * np.exp(A_M * year)
```

```python
ML_71 = Mooreslaw(1971)
ML_73 = Mooreslaw(1973)
ML_71, ML_73
```

```python
np.set_printoptions(suppress=True, formatter={'float_kind':'{:0.2f}'.format})
data = np.loadtxt("transistor_data.csv", delimiter=",", usecols=[1,2], skiprows=1)
```

```python

data
```

```python
years = data[:, 1]
transistor_counts = data[:, 0]
```

```python
print("year", years[:10])
print("transistor_counts", transistor_counts[:10])
```

```python
yi = np.log(transistor_counts)
```

```python
Z = years[:, np.newaxis] ** [1,0]
model = sm.OLS(yi, Z)
```

```python
results = model.fit()
print(results.summary())
```

```python
AB =results.params
```

```python
A = AB[0]
B = AB[1]
A,B
```

```python
print("Rate of semiconductors added on a chip every 2 years:")
print(
    "\tx{:.2f} +/- {:.2f} semiconductors per chip".format(
        np.exp((A) * 2), 2 * A * np.exp(2 * A) * 0.006
    )
)
```
